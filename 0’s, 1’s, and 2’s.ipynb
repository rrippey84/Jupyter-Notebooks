```python
def sort012(self, arr):
        start = 0
        current = 0
        end = len(arr) - 1

        while current <= end:
            if arr[current] == 0:
                arr[start], arr[current] = arr[current], arr[start]
                start += 1
                current += 1

            elif arr[current] == 1:
                current += 1

            else:  # arr[current] == 2
                arr[end], arr[current] = arr[current], arr[end]
                end -= 1

        return arr
```
